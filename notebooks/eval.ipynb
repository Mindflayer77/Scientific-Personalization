```python
import pandas as pd
import numpy as np

WR_COLS = ["wr_vs_base_wi", "wr_fs_vs_base_fs_wi", "wr_fs_vs_base_wi", "wr_vs_base_fs_wi"]
COL_NAMES = ["WR M1a vs B0", "WR M1b vs B1", "WR M1b vs B0", "WR M1a vs B1"]

PERSONA_ORDER = ["sota_chaser", "domain_expert", "rigorous_skeptic", "efficient_compute", "data_engineer"]
PERSONA_DISPLAY = {
    "sota_chaser":       "SOTA Chaser",
    "domain_expert":     "Domain Expert",
    "rigorous_skeptic":  "Rigorous Skeptic",
    "efficient_compute": "Efficient Compute",
    "data_engineer":     "Data Engineer",
}


def build_table(df):
    """Returns a DataFrame of 'mean ± std' strings, rows = personas + Overall."""
    grp = df.groupby("user_id")[WR_COLS]
    means = grp.mean().round(3).loc[PERSONA_ORDER]
    stds  = grp.std().round(3).loc[PERSONA_ORDER]
    means.index = [PERSONA_DISPLAY[p] for p in PERSONA_ORDER]
    stds.index  = means.index
    means.columns = COL_NAMES
    stds.columns  = COL_NAMES

    overall_mean = df[WR_COLS].mean().round(3)
    overall_std  = df[WR_COLS].std().round(3)
    overall_mean.index = COL_NAMES
    overall_std.index  = COL_NAMES

    # Combine into formatted strings
    cell_df = means.copy().astype(str)
    for col in COL_NAMES:
        for idx in means.index:
            cell_df.loc[idx, col] = f"{means.loc[idx, col]:.3f} $\\pm$ {stds.loc[idx, col]:.3f}"
        overall_str = f"{overall_mean[col]:.3f} $\\pm$ {overall_std[col]:.3f}"
        cell_df.loc["Łączny średni wynik", col] = overall_str

    # Store raw means for bold detection (including overall)
    all_means = means.copy()
    all_means.loc["Łączny średni wynik"] = overall_mean

    return cell_df, all_means


def to_latex(cell_df, all_means, caption, label):
    col_header = " & ".join(
        [f"\\textbf{{{c}}} $\\uparrow$" for c in COL_NAMES]
    )
    rows = []
    for i, (idx, row) in enumerate(cell_df.iterrows()):
        sep = "\\midrule\n" if idx == "Łączny średni wynik" else ""
        cells = []
        for col in COL_NAMES:
            val = row[col]
            if all_means.loc[idx, col] == all_means[col].max():
                val = f"\\textbf{{{val}}}"
            cells.append(val)
        rows.append(f"{sep}{idx} & {' & '.join(cells)} \\\\")

    body = "\n".join(rows)
    return f"""\\begin{{table}}[h]
\\centering
\\caption{{{caption}}}
\\label{{{label}}}
\\resizebox{{\\textwidth}}{{!}}{{%
\\begin{{tabular}}{{lcccc}}
\\toprule
\\textbf{{Persona}} & {col_header} \\\\
\\midrule
{body}
\\bottomrule
\\end{{tabular}}%
}}
\\end{{table}}"""


# ── Qwen3 ─────────────────────────────────────────────────────────────────────
df_qwen = pd.read_csv("experiment_1/judged-generate-dpo-qwen3-4B-full-gemma.csv")
cell_qwen, means_qwen = build_table(df_qwen)
print(to_latex(cell_qwen, means_qwen,
               caption="Wyniki Eksperymentu 1 -- Qwen3-4B",
               label="tab:exp1-qwen"))

print()

# ── Llama 3.2-3B ──────────────────────────────────────────────────────────────
df_llama = pd.read_csv("experiment_1/judged-generate-dpo-llama3.2-3B-full-gemma.csv")
cell_llama, means_llama = build_table(df_llama)
print(to_latex(cell_llama, means_llama,
               caption="Wyniki Eksperymentu 1 -- Llama3.2-3B",
               label="tab:exp1-llama"))
```

```python
import pandas as pd
import numpy as np

PERSONA_ORDER = ["sota_chaser", "domain_expert", "rigorous_skeptic", "efficient_compute", "data_engineer"]
PERSONA_DISPLAY = {
    "sota_chaser":       "SOTA Chaser",
    "domain_expert":     "Domain Expert",
    "rigorous_skeptic":  "Rigorous Skeptic",
    "efficient_compute": "Efficient Compute",
    "data_engineer":     "Data Engineer",
}

# Each metric: (display name, {config_label: column_name})
METRICS = {
    "Persona Adherence": {
        "B0":  "base_persona_adherence_score",
        "B1":  "base_fs_persona_adherence_score",
        "M1a": "trained_persona_adherence_score",
        "M1b": "trained_fs_persona_adherence_score",
    },
    "Groundedness": {
        "B0":  "base_groundedness_score",
        "B1":  "base_fs_groundedness_score",
        "M1a": "trained_groundedness_score",
        "M1b": "trained_fs_groundedness_score",
    },
    "Relevancy": {
        "B0":  "base_relevancy_score",
        "B1":  "base_fs_relevancy_score",
        "M1a": "trained_relevancy_score",
        "M1b": "trained_fs_relevancy_score",
    },
}

CONFIGS = ["B0", "B1", "M1a", "M1b"]


def build_metric_table(df, col_map):
    """Build cell_df + raw means for one metric across all configurations."""
    cols = [col_map[c] for c in CONFIGS]
    grp = df.groupby("user_id")[cols]
    means = grp.mean().round(3).loc[PERSONA_ORDER].copy()
    stds  = grp.std().round(3).loc[PERSONA_ORDER].copy()
    means.index = [PERSONA_DISPLAY[p] for p in PERSONA_ORDER]
    stds.index  = means.index
    means.columns = CONFIGS
    stds.columns  = CONFIGS

    overall_mean = df[cols].mean().round(3)
    overall_std  = df[cols].std().round(3)
    overall_mean.index = CONFIGS
    overall_std.index  = CONFIGS

    cell_df = means.copy().astype(str)
    for cfg in CONFIGS:
        for idx in means.index:
            cell_df.loc[idx, cfg] = (
                f"{means.loc[idx, cfg]:.3f} $\\pm$ {stds.loc[idx, cfg]:.3f}"
            )
        cell_df.loc["Łączny średni wynik", cfg] = (
            f"{overall_mean[cfg]:.3f} $\\pm$ {overall_std[cfg]:.3f}"
        )

    all_means = means.copy()
    all_means.loc["Łączny średni wynik"] = overall_mean
    return cell_df, all_means


def metric_to_latex(cell_df, all_means, caption, label):
    col_header = " & ".join([f"\\textbf{{{c}}}" for c in CONFIGS])
    rows = []
    for idx, row in cell_df.iterrows():
        sep = "\\midrule\n" if idx == "Łączny średni wynik" else ""
        cells = []
        for cfg in CONFIGS:
            val = row[cfg]
            if all_means.loc[idx, cfg] == all_means[cfg].max():
                val = f"\\textbf{{{val}}}"
            cells.append(val)
        rows.append(f"{sep}{idx} & {' & '.join(cells)} \\\\")
    body = "\n".join(rows)
    return f"""\\begin{{table}}[h]
\\centering
\\caption{{{caption}}}
\\label{{{label}}}
\\resizebox{{\\textwidth}}{{!}}{{%
\\begin{{tabular}}{{lcccc}}
\\toprule
\\textbf{{Persona}} & {col_header} \\\\
\\midrule
{body}
\\bottomrule
\\end{{tabular}}%
}}
\\end{{table}}"""


# ── Load both models ──────────────────────────────────────────────────────────
MODELS = {
    "Qwen3-4B":    "experiment_1/judged-generate-dpo-qwen3-4B-full-gemma.csv",
    "Llama3.2-3B": "experiment_1/judged-generate-dpo-llama3.2-3B-full-gemma.csv",
}

for model_name, csv_path in MODELS.items():
    df = pd.read_csv(csv_path)
    print(f"{'=' * 60}")
    print(f"  {model_name}")
    print(f"{'=' * 60}\n")
    for metric_name, col_map in METRICS.items():
        cell_df, all_means = build_metric_table(df, col_map)
        label = (f"tab:exp1-{model_name.lower().replace('.','').replace('-','')}"
                 f"-{metric_name.lower().replace(' ', '_')}")
        caption = f"Wyniki Eksperymentu 1 -- {model_name} -- {metric_name}"
        print(metric_to_latex(cell_df, all_means, caption=caption, label=label))
        print()
```

```python
import pandas as pd
import numpy as np

PERSONA_ORDER = ["sota_chaser", "domain_expert", "rigorous_skeptic", "efficient_compute", "data_engineer"]
PERSONA_DISPLAY = {
    "sota_chaser":       "SOTA Chaser",
    "domain_expert":     "Domain Expert",
    "rigorous_skeptic":  "Rigorous Skeptic",
    "efficient_compute": "Efficient Compute",
    "data_engineer":     "Data Engineer",
}
EXP3_PERSONA_FILE = {
    "sota_chaser":       "sota",
    "domain_expert":     "domain",
    "rigorous_skeptic":  "rigorous",
    "efficient_compute": "efficient",
    "data_engineer":     "data",
}
METRICS = {
    "Persona Adherence": ("trained_persona_adherence_score",    "trained_fs_persona_adherence_score"),
    "Groundedness":      ("trained_groundedness_score",         "trained_fs_groundedness_score"),
    "Relevancy":         ("trained_relevancy_score",            "trained_fs_relevancy_score"),
}
CONFIGS    = ["M1a", "M1b", "M3a", "M3b"]
CONFIG_HDR = {"M1a": r"$\mathrm{M1a}$", "M1b": r"$\mathrm{M1b}$",
              "M3a": r"$\mathrm{M3a}$", "M3b": r"$\mathrm{M3b}$"}
OVR = "Łączny średni wynik"


def load_exp3(model_suffix):
    frames = []
    for persona, fname in EXP3_PERSONA_FILE.items():
        df = pd.read_csv(f"experiment_3/judged-generate-cl-{fname}-{model_suffix}-gemma.csv")
        df["_persona"] = persona
        frames.append(df)
    return pd.concat(frames, ignore_index=True)


def build_comparison_table(df_exp1, df_exp3, trained_col, trained_fs_col):
    display_rows = [PERSONA_DISPLAY[p] for p in PERSONA_ORDER] + [OVR]
    # Use display names as keys directly
    rows_mean = {c: {} for c in CONFIGS}
    rows_std  = {c: {} for c in CONFIGS}

    grp1 = df_exp1.groupby("user_id")
    grp3 = df_exp3.groupby("_persona")

    for p in PERSONA_ORDER:
        dname = PERSONA_DISPLAY[p]
        for col, cfg in [(trained_col, "M1a"), (trained_fs_col, "M1b")]:
            s = grp1.get_group(p)[col]
            rows_mean[cfg][dname] = s.mean()
            rows_std [cfg][dname] = s.std()
        for col, cfg in [(trained_col, "M3a"), (trained_fs_col, "M3b")]:
            s = grp3.get_group(p)[col]
            rows_mean[cfg][dname] = s.mean()
            rows_std [cfg][dname] = s.std()

    for col, cfg in [(trained_col, "M1a"), (trained_fs_col, "M1b")]:
        rows_mean[cfg][OVR] = df_exp1[col].mean()
        rows_std [cfg][OVR] = df_exp1[col].std()
    for col, cfg in [(trained_col, "M3a"), (trained_fs_col, "M3b")]:
        rows_mean[cfg][OVR] = df_exp3[col].mean()
        rows_std [cfg][OVR] = df_exp3[col].std()

    means_df = pd.DataFrame(rows_mean, index=display_rows).round(3)
    stds_df  = pd.DataFrame(rows_std,  index=display_rows).round(3)

    cell_df = means_df.copy().astype(str)
    for cfg in CONFIGS:
        for idx in display_rows:
            cell_df.loc[idx, cfg] = (
                f"{means_df.loc[idx, cfg]:.3f} $\\pm$ {stds_df.loc[idx, cfg]:.3f}"
            )
    return cell_df, means_df


def comparison_to_latex(cell_df, means_df, caption, label):
    col_header = " & ".join([CONFIG_HDR[c] + r" $\uparrow$" for c in CONFIGS])
    rows = []
    for idx, row in cell_df.iterrows():
        sep = "\\midrule\n" if idx == OVR else ""
        cells = []
        for cfg in CONFIGS:
            val = row[cfg]
            if means_df.loc[idx, cfg] == means_df.loc[idx].max():
                val = f"\\textbf{{{val}}}"
            cells.append(val)
        rows.append(f"{sep}{idx} & {' & '.join(cells)} \\\\")
    body = "\n".join(rows)
    return f"""\\begin{{table}}[h]
\\centering
\\caption{{{caption}}}
\\label{{{label}}}
\\resizebox{{\\textwidth}}{{!}}{{%
\\begin{{tabular}}{{lcccc}}
\\toprule
\\textbf{{Persona}} & {col_header} \\\\
\\midrule
{body}
\\bottomrule
\\end{{tabular}}%
}}
\\end{{table}}"""


# ── Generate tables ───────────────────────────────────────────────────────────
MODELS = {
    "Qwen3-4B":    ("experiment_1/judged-generate-dpo-qwen3-4B-full-gemma.csv",    "qwen3-4B-full"),
    "Llama3.2-3B": ("experiment_1/judged-generate-dpo-llama3.2-3B-full-gemma.csv", "llama3.2-3B-full"),
}

for model_name, (exp1_path, exp3_suffix) in MODELS.items():
    df_exp1 = pd.read_csv(exp1_path)
    df_exp3 = load_exp3(exp3_suffix)

    print(f"{'=' * 60}")
    print(f"  {model_name}  (M1: wspólny adapter  |  M3: per-persona)")
    print(f"{'=' * 60}\n")

    for metric_name, (trained_col, trained_fs_col) in METRICS.items():
        cell_df, means_df = build_comparison_table(
            df_exp1, df_exp3, trained_col, trained_fs_col
        )
        slug  = model_name.lower().replace(".", "").replace("-", "")
        label = f"tab:exp1vs3-{slug}-{metric_name.lower().replace(' ', '_')}"
        caption = (
            f"Eksperyment 1 vs 3 -- {model_name} -- {metric_name} "
            r"(M1: wspólny adapter; M3: adapter per-persona)"
        )
        print(comparison_to_latex(cell_df, means_df, caption=caption, label=label))
        print()
```

```python
import pandas as pd
import numpy as np

# ── Config ────────────────────────────────────────────────────────────────────
PERSONA_ORDER = ["sota_chaser", "domain_expert", "rigorous_skeptic", "efficient_compute", "data_engineer"]
PERSONA_DISPLAY = {
    "sota_chaser":       "SOTA Chaser",
    "domain_expert":     "Domain Expert",
    "rigorous_skeptic":  "Rigorous Skeptic",
    "efficient_compute": "Efficient Compute",
    "data_engineer":     "Data Engineer",
}
OVR = "Łączny średni wynik"

# All 4 WR sub-columns — grouped: M_a together, M_b together
WR_RAW_COLS = [
    "wr_vs_base_wi",          # Mx_a vs B0
    "wr_vs_base_fs_wi",       # Mx_a vs B1
    "wr_fs_vs_base_wi",       # Mx_b vs B0
    "wr_fs_vs_base_fs_wi",    # Mx_b vs B1
]
WR_SUB_LABELS_DPO = [
    r"$M_{1a}$ vs $B_0$",
    r"$M_{1a}$ vs $B_1$",
    r"$M_{1b}$ vs $B_0$",
    r"$M_{1b}$ vs $B_1$",
]
WR_SUB_LABELS_SFT = [
    r"$M_{2a}$ vs $B_0$",
    r"$M_{2a}$ vs $B_1$",
    r"$M_{2b}$ vs $B_0$",
    r"$M_{2b}$ vs $B_1$",
]
# flat config keys: DPO first, SFT second
WR_CONFIGS = ["D1", "D2", "D3", "D4", "S1", "S2", "S3", "S4"]

METRICS = {
    "Persona Adherence": ("trained_persona_adherence_score", "trained_fs_persona_adherence_score"),
    "Groundedness":      ("trained_groundedness_score",      "trained_fs_groundedness_score"),
    "Relevancy":         ("trained_relevancy_score",         "trained_fs_relevancy_score"),
}
MET_CONFIGS = ["M1a", "M1b", "M2a", "M2b"]
MET_HDR = {c: rf"$\mathrm{{{c}}}$" for c in MET_CONFIGS}

MODELS = {
    "Qwen3-4B": {
        "dpo": "experiment_1/judged-generate-dpo-qwen3-4B-full-gemma.csv",
        "sft": "experiment_2/judged-generate-sft-qwen3-4B-full-gemma.csv",
    },
    "Llama3.2-3B": {
        "dpo": "experiment_1/judged-generate-dpo-llama3.2-3B-full-gemma.csv",
        "sft": "experiment_2/judged-generate-sft-llama3.2-3B-full-gemma.csv",
    },
}


def _display_rows():
    return [PERSONA_DISPLAY[p] for p in PERSONA_ORDER] + [OVR]


def _fmt(mean, std):
    return f"{mean:.3f} $\\pm$ {std:.3f}"


# ── Win-rate comparison (8-column, multicolumn header) ────────────────────────
def build_wr_comparison(df_dpo, df_sft):
    drows = _display_rows()
    rm = {c: {} for c in WR_CONFIGS}
    rs = {c: {} for c in WR_CONFIGS}
    g_dpo = df_dpo.groupby("user_id")
    g_sft = df_sft.groupby("user_id")

    # Per-persona: mean only (no std)
    for p in PERSONA_ORDER:
        dn = PERSONA_DISPLAY[p]
        for i, raw in enumerate(WR_RAW_COLS):
            rm[f"D{i+1}"][dn] = g_dpo.get_group(p)[raw].mean()
            rm[f"S{i+1}"][dn] = g_sft.get_group(p)[raw].mean()

    # Overall: mean of per-persona means; std = std of per-persona means (ddof=1)
    for i, raw in enumerate(WR_RAW_COLS):
        pp_dpo = [g_dpo.get_group(p)[raw].mean() for p in PERSONA_ORDER]
        pp_sft = [g_sft.get_group(p)[raw].mean() for p in PERSONA_ORDER]
        rm[f"D{i+1}"][OVR] = np.mean(pp_dpo)
        rs[f"D{i+1}"][OVR] = np.std(pp_dpo, ddof=1)
        rm[f"S{i+1}"][OVR] = np.mean(pp_sft)
        rs[f"S{i+1}"][OVR] = np.std(pp_sft, ddof=1)

    md = pd.DataFrame(rm, index=drows).round(3)
    sd = pd.DataFrame(rs, index=drows).round(3)

    persona_names = [PERSONA_DISPLAY[p] for p in PERSONA_ORDER]
    cd = md.copy().astype(str)
    for cfg in WR_CONFIGS:
        for idx in persona_names:
            cd.loc[idx, cfg] = f"{md.loc[idx, cfg]:.3f}"
        cd.loc[OVR, cfg] = _fmt(md.loc[OVR, cfg], sd.loc[OVR, cfg])
    return cd, md


def wr_to_latex(cell_df, means_df, caption, label):
    sub_dpo = " & ".join([f"{l} $\\uparrow$" for l in WR_SUB_LABELS_DPO])
    sub_sft = " & ".join([f"{l} $\\uparrow$" for l in WR_SUB_LABELS_SFT])
    col_header = (
        r"\textbf{Persona}"
        r" & \multicolumn{4}{c}{\textbf{SFT+DPO}}"
        r" & \multicolumn{4}{c}{\textbf{SFT}} \\"
        "\n"
        r"\cmidrule(lr){2-5} \cmidrule(lr){6-9}"
        "\n"
        f" & {sub_dpo} & {sub_sft}"
    )
    rows = []
    for idx, row in cell_df.iterrows():
        sep = "\\midrule\n" if idx == OVR else ""
        row_max = means_df.loc[idx].max()
        cells = []
        for cfg in WR_CONFIGS:
            val = row[cfg]
            if means_df.loc[idx, cfg] == row_max:
                val = f"\\textbf{{{val}}}"
            cells.append(val)
        rows.append(f"{sep}{idx} & {' & '.join(cells)} \\\\")
    body = "\n".join(rows)
    return (
        f"\\begin{{table}}[h]\n\\centering\n\\caption{{{caption}}}\n\\label{{{label}}}\n"
        f"\\resizebox{{\\textwidth}}{{!}}{{%\n\\begin{{tabular}}{{lcccccccc}}\n\\toprule\n"
        f"{col_header} \\\\\n\\midrule\n{body}\n"
        f"\\bottomrule\n\\end{{tabular}}%\n}}\n\\end{{table}}"
    )


# ── Metric comparison (4-column) ──────────────────────────────────────────────
def build_met_comparison(df_dpo, df_sft, col_a, col_b):
    drows = _display_rows()
    col_map = {"M1a": (df_dpo, col_a), "M1b": (df_dpo, col_b),
               "M2a": (df_sft, col_a), "M2b": (df_sft, col_b)}
    rm = {c: {} for c in MET_CONFIGS}
    rs = {c: {} for c in MET_CONFIGS}
    grps = {k: v[0].groupby("user_id") for k, v in col_map.items()}
    for p in PERSONA_ORDER:
        dn = PERSONA_DISPLAY[p]
        for cfg, (df, col) in col_map.items():
            s = grps[cfg].get_group(p)[col]
            rm[cfg][dn] = s.mean(); rs[cfg][dn] = s.std()
    for cfg, (df, col) in col_map.items():
        rm[cfg][OVR] = df[col].mean(); rs[cfg][OVR] = df[col].std()
    md = pd.DataFrame(rm, index=drows).round(3)
    sd = pd.DataFrame(rs, index=drows).round(3)
    cd = md.copy().astype(str)
    for cfg in MET_CONFIGS:
        for idx in drows:
            cd.loc[idx, cfg] = _fmt(md.loc[idx, cfg], sd.loc[idx, cfg])
    return cd, md


def met_to_latex(cell_df, means_df, caption, label):
    col_header = (
        r"\textbf{Persona}"
        r" & \multicolumn{2}{c}{\textbf{DPO (Eksp. 1)}}"
        r" & \multicolumn{2}{c}{\textbf{SFT (Eksp. 2)}} \\"
        "\n"
        r"\cmidrule(lr){2-3} \cmidrule(lr){4-5}"
        "\n"
        r" & $\mathrm{M1a}$ $\uparrow$ & $\mathrm{M1b}$ $\uparrow$"
        r" & $\mathrm{M2a}$ $\uparrow$ & $\mathrm{M2b}$ $\uparrow$"
    )
    rows = []
    for idx, row in cell_df.iterrows():
        sep = "\\midrule\n" if idx == OVR else ""
        row_max = means_df.loc[idx].max()
        cells = []
        for cfg in MET_CONFIGS:
            val = row[cfg]
            if means_df.loc[idx, cfg] == row_max:
                val = f"\\textbf{{{val}}}"
            cells.append(val)
        rows.append(f"{sep}{idx} & {' & '.join(cells)} \\\\")
    body = "\n".join(rows)
    return (
        f"\\begin{{table}}[h]\n\\centering\n\\caption{{{caption}}}\n\\label{{{label}}}\n"
        f"\\resizebox{{\\textwidth}}{{!}}{{%\n\\begin{{tabular}}{{lcccc}}\n\\toprule\n"
        f"{col_header} \\\\\n\\midrule\n{body}\n"
        f"\\bottomrule\n\\end{{tabular}}%\n}}\n\\end{{table}}"
    )


# ── Generate all tables ───────────────────────────────────────────────────────
for model_name, paths in MODELS.items():
    df_dpo = pd.read_csv(paths["dpo"])
    df_sft = pd.read_csv(paths["sft"])
    slug = model_name.lower().replace(".", "").replace("-", "")

    print(f"{'=' * 60}")
    print(f"  {model_name}  (SFT+DPO Eksp. 1  |  SFT Eksp. 2)")
    print(f"{'=' * 60}\n")

    # Win rate
    cd, md = build_wr_comparison(df_dpo, df_sft)
    print(wr_to_latex(cd, md,
                      caption=f"Współczynnik wygranych -- SFT+DPO (E1) vs SFT (E2) -- {model_name}",
                      label=f"tab:wr-exp1vs2-{slug}"))
    print()

    # Metrics
    for metric_name, (col_a, col_b) in METRICS.items():
        cd, md = build_met_comparison(df_dpo, df_sft, col_a, col_b)
        print(met_to_latex(cd, md,
                           caption=f"DPO (Eksp. 1) vs SFT (Eksp. 2) -- {model_name} -- {metric_name}",
                           label=f"tab:met-exp1vs2-{slug}-{metric_name.lower().replace(' ', '_')}"))
        print()
```

## Porównanie długości kontekstu: final_2/dpo_regen vs final_3/final_dpo-2

Porównanie długości kontekstu (kolumna `context`) między dwoma próbkami danych, zmierzone tokenizatorami dla modeli **Qwen3-4B** i **Llama3.2-3B-Instruct**.

- **Próbka 1**: `hypotheses/final_2/dpo_regen.csv` – pełny zbiór (50 przykładów, 10 na personę)
- **Próbka 2**: `hypotheses/final_3/final_dpo-2.csv` – losowanie stratyfikowane 50 przykładów (10 na personę wg `chosen_persona`)

```python
from transformers import AutoTokenizer

QWEN3_MODEL = "Qwen/Qwen3-4B"
LLAMA_MODEL = "meta-llama/Llama-3.2-3B-Instruct"

print("Ładowanie tokenizatora Qwen3-4B...")
qwen_tok = AutoTokenizer.from_pretrained(QWEN3_MODEL)
print("Ładowanie tokenizatora Llama3.2-3B-Instruct...")
llama_tok = AutoTokenizer.from_pretrained(LLAMA_MODEL)
print("Tokenizatory załadowane.")
```

```python
import pandas as pd

# Próbka 1: hypotheses/final_2/dpo_regen.csv (50 przykładów, już zbalansowane)
df1 = pd.read_csv("../final_2/data_eval/dpo-gemma.csv")
df1 = df1[df1['hypothesis_type'] == 'chosen']
print(f"Próbka 1 – kształt: {df1.shape}")
print(df1["persona"].value_counts())

print()

# Próbka 2: hypotheses/final_3/final_dpo-2.csv – stratyfikowane losowanie 50 przykładów (10 na personę)
df2_full = pd.read_csv("data_eval/eval-dpo-gemma.csv")
# df2 = (
#     df2_full
#     .groupby("persona", group_keys=False)
#     .apply(lambda g: g.sample(n=10, random_state=42))
#     .reset_index(drop=True)
# )
df2 = df2_full[df2_full['hypothesis_type'] == 'chosen']
print(f"Próbka 2 – kształt: {df2.shape}")
print(df2["persona"].value_counts())
```

```python
def count_tokens(tokenizer, texts):
    """Zwraca listę liczby tokenów dla każdego tekstu (bez specjalnych tokenów)."""
    return [len(tokenizer.encode(str(t), add_special_tokens=False)) for t in texts]

for label, df in [("Próbka 1 (dpo_regen)", df1), ("Próbka 2 (final_dpo-2)", df2)]:
    df[f"ctx_tokens_qwen3"] = count_tokens(qwen_tok, df["context"])
    df[f"ctx_tokens_llama"] = count_tokens(llama_tok, df["context"])
    print(f"\n{label}")
    print(df[[f"ctx_tokens_qwen3", f"ctx_tokens_llama"]].describe().round(1))
```

## Analiza Length Bias (Eksperyment 1)

Czy sędzia LLM preferuje dłuższe hipotezy? Dla każdego porównania (M1a vs B0, M1a vs B1, M1b vs B0, M1b vs B1) obliczamy:
- średnią długość tekstu (znaki, słowa) dla wygrywającej i przegrywającej strony,
- % przypadków, gdy dłuższa hipoteza wygrała,
- korelację punkt-biserialną między różnicą długości a wynikiem (win=1).


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
from scipy import stats

# ── Config ────────────────────────────────────────────────────────────────────
# (label, wi_col, col_A, col_B, A_name, B_name)
# wi_col == 1.0 means A won, 0.0 means B won
COMPARISONS = [
    ("M1a vs B0", "wr_vs_base_wi",       "trained_response",    "base_response",    "M1a", "B0"),
    ("M1a vs B1", "wr_vs_base_fs_wi",    "trained_response",    "base_fs_response", "M1a", "B1"),
    ("M1b vs B0", "wr_fs_vs_base_wi",    "trained_fs_response", "base_response",    "M1b", "B0"),
    ("M1b vs B1", "wr_fs_vs_base_fs_wi", "trained_fs_response", "base_fs_response", "M1b", "B1"),
]

MODELS = {
    "Qwen3-4B": {
        "gen":    "experiment_1/generate-dpo-10-qwen3-4B-full.csv",
        "judged": "experiment_1/judged-generate-dpo-qwen3-4B-full-gemma.csv",
    },
    "Llama3.2-3B": {
        "gen":    "experiment_1/generate-dpo-llama3.2-3B-full.csv",
        "judged": "experiment_1/judged-generate-dpo-llama3.2-3B-full-gemma.csv",
    },
}

PERSONA_ORDER = ["sota_chaser", "domain_expert", "rigorous_skeptic", "efficient_compute", "data_engineer"]
PERSONA_DISPLAY = {
    "sota_chaser":       "SOTA Chaser",
    "domain_expert":     "Domain Expert",
    "rigorous_skeptic":  "Rigorous Skeptic",
    "efficient_compute": "Efficient Compute",
    "data_engineer":     "Data Engineer",
}


def _chars(text):
    return len(str(text)) if pd.notna(text) else 0


def _words(text):
    return len(str(text).split()) if pd.notna(text) else 0


def analyze_length_bias(df):
    """Per-comparison length bias stats. Returns a DataFrame."""
    rows = []
    for comp_label, wi_col, col_a, col_b, a_name, b_name in COMPARISONS:
        sub = df[[wi_col, col_a, col_b, "user_id"]].dropna()
        len_a = sub[col_a].apply(_chars)
        len_b = sub[col_b].apply(_chars)
        len_a_w = sub[col_a].apply(_words)
        len_b_w = sub[col_b].apply(_words)

        won_a = sub[wi_col] == 1.0  # A wins
        won_b = sub[wi_col] == 0.0  # B wins

        diff_chars = len_a - len_b
        diff_words = len_a_w - len_b_w

        # 1 if longer side won, 0 if shorter side won, 0.5 if equal
        longer_wins = np.where(
            diff_chars > 0, won_a.astype(float),
            np.where(diff_chars < 0, won_b.astype(float), 0.5)
        )
        pct_longer_wins = longer_wins.mean() * 100

        corr_c, pval_c = stats.pointbiserialr(sub[wi_col], diff_chars)
        corr_w, pval_w = stats.pointbiserialr(sub[wi_col], diff_words)

        rows.append({
            "Porównanie":                comp_label,
            "n":                         len(sub),
            f"Śr. dl. {a_name} (zn.) [wygr.]":  round(len_a[won_a].mean(), 0),
            f"Śr. dl. {a_name} (zn.) [prz.]":   round(len_a[won_b].mean(), 0),
            f"Śr. dl. {b_name} (zn.) [wygr.]":  round(len_b[won_b].mean(), 0),
            f"Śr. dl. {b_name} (zn.) [prz.]":   round(len_b[won_a].mean(), 0),
            "% dłuższa wygrywa":          round(pct_longer_wins, 1),
            "r (Δznaki, win)":            round(corr_c, 3),
            "p (Δznaki)":                 round(pval_c, 4),
            "r (Δsłowa, win)":            round(corr_w, 3),
            "p (Δsłowa)":                 round(pval_w, 4),
        })
    return pd.DataFrame(rows).set_index("Porównanie")


def analyze_length_bias_by_persona(df):
    """Per-persona × per-comparison % dłuższa wygrywa."""
    records = []
    for comp_label, wi_col, col_a, col_b, a_name, b_name in COMPARISONS:
        for persona in PERSONA_ORDER:
            sub = df[df["user_id"] == persona][[wi_col, col_a, col_b]].dropna()
            if len(sub) == 0:
                continue
            len_a = sub[col_a].apply(_chars)
            len_b = sub[col_b].apply(_chars)
            diff = len_a - len_b
            won_a = sub[wi_col] == 1.0
            won_b = sub[wi_col] == 0.0
            longer_wins = np.where(
                diff > 0, won_a.astype(float),
                np.where(diff < 0, won_b.astype(float), 0.5)
            )
            records.append({
                "Porównanie": comp_label,
                "Persona":    PERSONA_DISPLAY[persona],
                "n":          len(sub),
                "% dłuższa wygrywa": round(longer_wins.mean() * 100, 1),
                "Śr. Δ znaków (A−B)": round(diff.mean(), 0),
            })
    return pd.DataFrame(records)


# ── Load & merge ──────────────────────────────────────────────────────────────
merged_lb = {}

for model_name, paths in MODELS.items():
    df_gen    = pd.read_csv(paths["gen"])
    df_judged = pd.read_csv(paths["judged"])

    merge_keys = ["user_id", "question"]
    text_cols  = ["trained_response", "trained_fs_response",
                  "base_response",    "base_fs_response"]
    df = df_judged.merge(df_gen[merge_keys + text_cols], on=merge_keys, how="left")
    merged_lb[model_name] = df

# ── Print overall stats ───────────────────────────────────────────────────────
for model_name, df in merged_lb.items():
    print(f"\n{'='*70}")
    print(f"  {model_name}  –  Length Bias (znaki)")
    print(f"{'='*70}")
    result = analyze_length_bias(df)
    print(result[["n", "% dłuższa wygrywa", "r (Δznaki, win)", "p (Δznaki)",
                  "r (Δsłowa, win)", "p (Δsłowa)"]].to_string())
```

```python
import pandas as pd
import numpy as np

# ── Config ────────────────────────────────────────────────────────────────────
PERSONA_ORDER = ["sota_chaser", "domain_expert", "rigorous_skeptic", "efficient_compute", "data_engineer"]
PERSONA_DISPLAY = {
    "sota_chaser":       "SOTA Chaser",
    "domain_expert":     "Domain Expert",
    "rigorous_skeptic":  "Rigorous Skeptic",
    "efficient_compute": "Efficient Compute",
    "data_engineer":     "Data Engineer",
}
EXP_PERSONA_FILE = {
    "sota_chaser":       "sota",
    "domain_expert":     "domain",
    "rigorous_skeptic":  "rigorous",
    "efficient_compute": "efficient",
    "data_engineer":     "data",
}
METRICS = {
    "Persona Adherence": ("trained_persona_adherence_score", "trained_fs_persona_adherence_score"),
    "Groundedness":      ("trained_groundedness_score",       "trained_fs_groundedness_score"),
    "Relevancy":         ("trained_relevancy_score",          "trained_fs_relevancy_score"),
}
CONFIGS    = ["M3a", "M3b", "M4a"]
CONFIG_HDR = {c: rf"$\mathrm{{{c}}}$" for c in CONFIGS}
OVR = "Łączny średni wynik"


def load_exp3(model_suffix):
    frames = []
    for persona, fname in EXP_PERSONA_FILE.items():
        df = pd.read_csv(f"experiment_3/judged-generate-cl-{fname}-{model_suffix}-gemma.csv")
        df = df[df["user_id"] == persona]  # filter to only this persona's rows
        df["_persona"] = persona
        frames.append(df)
    return pd.concat(frames, ignore_index=True)


def load_exp5(model_suffix):
    frames = []
    for persona, fname in EXP_PERSONA_FILE.items():
        df = pd.read_csv(f"experiment_5/judged-cl-{fname}-rationale-{model_suffix}.csv")
        df["_persona"] = persona
        frames.append(df)
    return pd.concat(frames, ignore_index=True)


def build_comparison_table(df_exp3, df_exp5, trained_col, trained_fs_col):
    display_rows = [PERSONA_DISPLAY[p] for p in PERSONA_ORDER] + [OVR]
    rows_mean = {c: {} for c in CONFIGS}
    rows_std  = {c: {} for c in CONFIGS}
    grp3 = df_exp3.groupby("_persona")
    grp5 = df_exp5.groupby("_persona")
    for p in PERSONA_ORDER:
        dn = PERSONA_DISPLAY[p]
        s3a = grp3.get_group(p)[trained_col]
        s3b = grp3.get_group(p)[trained_fs_col]
        s5a = grp5.get_group(p)[trained_col]
        rows_mean["M3a"][dn] = s3a.mean(); rows_std["M3a"][dn] = s3a.std()
        rows_mean["M3b"][dn] = s3b.mean(); rows_std["M3b"][dn] = s3b.std()
        rows_mean["M4a"][dn] = s5a.mean(); rows_std["M4a"][dn] = s5a.std()
    rows_mean["M3a"][OVR] = df_exp3[trained_col].mean();    rows_std["M3a"][OVR] = df_exp3[trained_col].std()
    rows_mean["M3b"][OVR] = df_exp3[trained_fs_col].mean(); rows_std["M3b"][OVR] = df_exp3[trained_fs_col].std()
    rows_mean["M4a"][OVR] = df_exp5[trained_col].mean();    rows_std["M4a"][OVR] = df_exp5[trained_col].std()
    means_df = pd.DataFrame(rows_mean, index=display_rows).round(3)
    stds_df  = pd.DataFrame(rows_std,  index=display_rows).round(3)
    cell_df = means_df.copy().astype(str)
    for cfg in CONFIGS:
        for idx in display_rows:
            cell_df.loc[idx, cfg] = (
                f"{means_df.loc[idx, cfg]:.3f} $\\pm$ {stds_df.loc[idx, cfg]:.3f}"
            )
    return cell_df, means_df


def comparison_to_latex(cell_df, means_df, caption, label):
    col_header = " & ".join([CONFIG_HDR[c] + r" $\uparrow$" for c in CONFIGS])
    rows = []
    for idx, row in cell_df.iterrows():
        sep = "\\midrule\n" if idx == OVR else ""
        cells = []
        for cfg in CONFIGS:
            val = row[cfg]
            if means_df.loc[idx, cfg] == means_df.loc[idx].max():
                val = f"\\textbf{{{val}}}"
            cells.append(val)
        rows.append(f"{sep}{idx} & {' & '.join(cells)} \\\\")
    body = "\n".join(rows)
    return (
        f"\\begin{{table}}[h]\n"
        f"\\centering\n"
        f"\\caption{{{caption}}}\n"
        f"\\label{{{label}}}\n"
        f"\\resizebox{{\\textwidth}}{{!}}{{%\n"
        f"\\begin{{tabular}}{{lccc}}\n"
        f"\\toprule\n"
        f"\\textbf{{Persona}} & {col_header} \\\\\n"
        f"\\midrule\n"
        f"{body}\n"
        f"\\bottomrule\n"
        f"\\end{{tabular}}%\n"
        f"}}\n"
        f"\\end{{table}}"
    )


# ── Models ────────────────────────────────────────────────────────────────────
MODELS = {
    "Qwen3-4B":    ("qwen3-4B-full", "qwen"),
    "Llama3.2-3B": ("llama3.2-3B-full", "llama"),
}

for model_name, (exp3_suffix, exp5_suffix) in MODELS.items():
    df_exp3 = load_exp3(exp3_suffix)
    df_exp5 = load_exp5(exp5_suffix)

    print(f"{'=' * 60}")
    print(f"  {model_name}  (M3: per-persona CL  |  M5: per-persona CL + rationale)")
    print(f"{'=' * 60}\n")

    for metric_name, (trained_col, trained_fs_col) in METRICS.items():
        cell_df, means_df = build_comparison_table(
            df_exp3, df_exp5, trained_col, trained_fs_col
        )
        slug  = model_name.lower().replace(".", "").replace("-", "")
        label = f"tab:exp3vs4-{slug}-{metric_name.lower().replace(' ', '_')}"
        caption = (
            f"Eksperyment 3 vs 4 -- {model_name} -- {metric_name} "
            r"(M3: per-persona CL; M5: per-persona CL + rationale)"
        )
        print(comparison_to_latex(cell_df, means_df, caption=caption, label=label))
        print()
```

```python
import pandas as pd
import numpy as np

PERSONA_ORDER = ["sota_chaser", "domain_expert", "rigorous_skeptic", "efficient_compute", "data_engineer"]
PERSONA_DISPLAY = {
    "sota_chaser":       "SOTA Chaser",
    "domain_expert":     "Domain Expert",
    "rigorous_skeptic":  "Rigorous Skeptic",
    "efficient_compute": "Efficient Compute",
    "data_engineer":     "Data Engineer",
}
OVR = "\u0141\u0105czny \u015bredni wynik"

# Column order matching the original table: Ma vs B0, Mb vs B1, Mb vs B0, Ma vs B1
WR_COLS = [
    "wr_vs_base_wi",
    "wr_vs_base_fs_wi",
    "wr_fs_vs_base_wi",
    "wr_fs_vs_base_fs_wi",
]
COL_LABELS_DPO = [
    r"$M_{1a}$ vs $B_0$",
    r"$M_{1a}$ vs $B_1$",
    r"$M_{1b}$ vs $B_0$",
    r"$M_{1b}$ vs $B_1$",
]
COL_LABELS_SFT = [
    r"$M_{2a}$ vs $B_0$",
    r"$M_{2a}$ vs $B_1$",
    r"$M_{2b}$ vs $B_0$",
    r"$M_{2b}$ vs $B_1$",
]

df_dpo = pd.read_csv("experiment_1/judged-generate-dpo-llama3.2-3B-full-gemma.csv")
df_sft = pd.read_csv("experiment_2/judged-generate-sft-llama3.2-3B-full-gemma.csv")

def compute_wr(df):
    g = df.groupby("user_id")
    pp = {PERSONA_DISPLAY[p]: np.array([g.get_group(p)[c].mean() for c in WR_COLS])
          for p in PERSONA_ORDER}
    arr = np.array([pp[PERSONA_DISPLAY[p]] for p in PERSONA_ORDER])
    return pp, arr.mean(axis=0), arr.std(axis=0, ddof=1)

pp_dpo, om_dpo, os_dpo = compute_wr(df_dpo)
pp_sft, om_sft, os_sft = compute_wr(df_sft)

def bold(cell, flag):
    return f"\\textbf{{{cell}}}" if flag else cell

def make_row(label, dpo_vals, sft_vals, dpo_cells, sft_cells):
    all_vals = np.concatenate([dpo_vals, sft_vals])
    row_max = all_vals.max()
    dpo_out = [bold(c, abs(v - row_max) < 1e-9) for c, v in zip(dpo_cells, dpo_vals)]
    sft_out = [bold(c, abs(v - row_max) < 1e-9) for c, v in zip(sft_cells, sft_vals)]
    return label + " & " + " & ".join(dpo_out + sft_out) + " \\\\"

rows = []
for p in PERSONA_ORDER:
    dn = PERSONA_DISPLAY[p]
    dv, sv = pp_dpo[dn], pp_sft[dn]
    rows.append(make_row(dn, dv, sv,
                         [f"{v:.3f}" for v in dv],
                         [f"{v:.3f}" for v in sv]))

rows.append("\\midrule")
rows.append(make_row(OVR, om_dpo, om_sft,
                     [f"{m:.3f} $\\pm$ {s:.3f}" for m, s in zip(om_dpo, os_dpo)],
                     [f"{m:.3f} $\\pm$ {s:.3f}" for m, s in zip(om_sft, os_sft)]))

sub_hdr_dpo = " & ".join([f"{l} $\\uparrow$" for l in COL_LABELS_DPO])
sub_hdr_sft = " & ".join([f"{l} $\\uparrow$" for l in COL_LABELS_SFT])

lines = [
    "\\begin{table}[h]",
    "\\centering",
    "\\caption{Wsp\u00f3\u0142czynnik wygranych -- SFT+DPO (E1) vs SFT (E2) -- Llama3.2-3B}",
    "\\label{tab:wr-exp1vs2-llama323b}",
    "\\resizebox{\\textwidth}{!}{%",
    "\\begin{tabular}{lcccccccc}",
    "\\toprule",
    "\\textbf{Persona} & \\multicolumn{4}{c}{\\textbf{SFT+DPO}} & \\multicolumn{4}{c}{\\textbf{SFT}} \\\\",
    "\\cmidrule(lr){2-5} \\cmidrule(lr){6-9}",
    f" & {sub_hdr_dpo} & {sub_hdr_sft} \\\\",
    "\\midrule",
] + rows + [
    "\\bottomrule",
    "\\end{tabular}%",
    "}",
    "\\end{table}",
]
print("\n".join(lines))
```
